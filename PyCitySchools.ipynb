```python
#Calculate percentage of students that passed math
passed_math = (merged_data[merged_data['math_score']>=65]['gender'].count()/unique_student_names)*100

#preview
passed_math
```

```python
#Calculate percentage of students that passed reading
passed_reading = (merged_data[merged_data['reading_score']>=65]['gender'].count()/unique_student_names)*100

#Preview
passed_reading
```

```python
#Calculate overall passing percentage
passed_both = (merged_data[(merged_data['math_score']>=65) & (merged_data['reading_score']>=65)]['gender'].count()/unique_student_names)*100

#Preview
passed_both
```
